```python
# The interpretation and computation of entropy
'''
H = - sum[ p(xi)*log(p(xi))] for all possible events
'''
import requests
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://www.gutenberg.org/files/35/35-0.txt'
r = requests.get(url, allow_redirects=True)

open('TheTimeMachine.txt', 'wb').write(r.content)
```

```python
book = requests.get(url)
text = book.text
print(type(text))
print(len(text))
#text[:2000]
```

```python
# character strings to replace with space
strings2replace = [
                 '\r\n\r\nâ\x80\x9c', # new paragraph
                 'â\x80\x9c',         # open quote
                 'â\x80\x9d',         # close quote
                 '\r\n',              # new line
                 'â\x80\x94',         # hyphen
                 'â\x80\x99',         # single apostrophe
                 'â\x80\x98',         # single quote
                 '_',                 # underscore, used for stressing
                 'ï»¿',
    ]
```

```python
for str2match in strings2replace:
    regexp = re.compile('%s'%str2match)
    text = regexp.sub(' ', text)
text = text.strip()
text[:100]
```

```python
words = text.split()
print(type(words))
print(len(words))
```

```python
import numpy as np
import matplotlib.pyplot as plt
#words[:50]
wordlengths = np.zeros(len(words))
for wordi in range(len(words)):
    wordlengths[wordi] = len(words[wordi])
```

```python
plt.plot(wordlengths, 'o')
plt.show()
```

```python
plt.hist(wordlengths, bins=30)
plt.xlabel('Word lenghts')
plt.ylabel('word counts')
plt.show()
```

```python
wordcounts = np.zeros(40)
for wrd in words:
    wordcounts[len(wrd)] += 1
#print(wordcounts)

plt.bar(range(len(wordcounts)), wordcounts)
plt.show()

#plt.hist(wordcounts, bins=40)
#plt.xlabel('Word lenghts')
#plt.ylabel('word counts')
#plt.show()
```

```python
# how to get alphabet in python
import string
alphabet_string = string.ascii_lowercase
#alphabet_string
alphadict = dict.fromkeys(alphabet_string, 0)
#alphadict

for char in text.lower():
    if char not in alphabet_string:
        continue
    #print(char)
    alphadict[char] += 1
```

```python
items = alphadict.items()
x,y = zip(*items)
#plt.plot(x,y)
plt.bar(x,y)
plt.show()
```

```python
letters = string.ascii_lowercase
text.lower().count('a')
```

```python
lettercounts = np.zeros(len(letters))
n = 0
for i in letters:
    #lettercounts[letters.index(i)] = text.lower().count((i))
    lettercounts[n] = text.lower().count((i))
    n += 1
lettercounts
```

```python
plt.bar(list(letters),lettercounts)
plt.show()
```

```python

numletters = len(letters)
fig, ax = plt.subplots(1, figsize=(15,4))
ax.bar(range(numletters),lettercounts)
ax.set_xticks(range(numletters))
ax.set_xticklabels(letters)
ax.set_xlabel('Letter')
ax.set_ylabel('Letter Count')
plt.show()
```

```python
letterprob = lettercounts/sum(lettercounts)

fig, ax = plt.subplots(1, figsize=(15,4))
ax.bar(range(numletters),letterprob)
ax.set_xticks(range(numletters))
ax.set_xticklabels(letters)
ax.set_xlabel('Letter')
ax.set_ylabel('Probability')
#plt.show()
fig;
```

```python
#epsilon = 0.0000001
entropy = -sum(letterprob * np.log2(letterprob + np.finfo(float).eps))
entropy
```

```python
np.log2(0 + np.finfo(float).eps)
```

```python
ax.set_title('Entropy=%.3f'%entropy)
fig
```

```python
# Conditional Entropy
# Sequential ?
probmat = np.zeros((numletters, numletters))
#probmat.shape
for i in range(len(text)-1):
    currlet = text[i]
    nextlet = text[i+1]
    if currlet in letters and nextlet in letters:
        probmat[letters.index(currlet), letters.index(nextlet)] += 1
```

```python
fig,ax = plt.subplots(1, figsize=(6,6))
ax.imshow(probmat, vmax=500)
ax.set_ylabel('Current letter')
ax.set_xlabel('Next letter')
ax.set_yticks(range(numletters))
ax.set_xticks(range(numletters))
ax.set_xticklabels(letters)
ax.set_yticklabels(letters)
plt.show()
```

```python
condentr = np.zeros(numletters)
for i in range(numletters):
    probs = probmat[i,:]
    probs = probs/sum(probs)
    condentr[i] = -sum( probs*np.log2(probs + np.finfo(float).eps))
#condentr
```

```python
fig,ax = plt.subplots(1, figsize=(8,5))
ax.bar(list(letters), condentr)
plt.show()
```
